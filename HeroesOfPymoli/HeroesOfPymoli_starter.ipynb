```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
# Find the number of unique items in the data set
unique_items = purchase_data['Item Name'].unique()
unique = len(unique_items)
# Use Pandas to calculate the average price of items and format to currency
avprice = purchase_data['Price'].mean()
avpricef = "${:,.2f}".format(avprice)
# Find the number of purchases in the data set
purchases = purchase_data['Purchase ID'].value_counts()
numpur = len(purchases)
# Use pandas to calculate the total revenue and format to currency
totalrev = purchase_data['Price'].sum()
totalrevf = "${:,.2f}".format(totalrev)
# Create summary table with above information
summary = pd.DataFrame(
    {"Number of Unique Items":[unique],
    "Average Price":[avpricef],
    "Number of Purchases":[numpur],
    "Total Revenue":[totalrevf]})
summary.head()
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
# Drop duplicate SN rows for accurate count of users
unique_sn = purchase_data.drop_duplicates(subset=['SN'])
# Create a dataframe from the total counts of users by gender
gender_sum = unique_sn['Gender'].value_counts().rename_axis('Gender').reset_index(name='Total Count')
# Add a column to the dataframe of the calculated percentage of users by gender
gender_sum['Percentage of Players'] = gender_sum['Total Count']/gender_sum['Total Count'].sum()*100
# Reset the index to the Gender column
gender_sum = gender_sum.set_index('Gender')
# Convert the Percentage of Players column from string to numeric
gender_sum['Percentage of Players'] = pd.to_numeric(gender_sum['Percentage of Players'])
# Format the Percentage of Players column to show as a percent
gender_sum['Percentage of Players'] = pd.Series(["{0:.2f}%".format(val) for val in gender_sum['Percentage of Players']], index=gender_sum.index)
gender_sum.head()
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Create new dataframe summary table from Gender Values in data set
gender_calc = purchase_data['Gender'].value_counts().rename_axis('Gender').reset_index(name='Purchase Count')
# Reset the index to Gender column
gender_calc = gender_calc.set_index('Gender')
# Create a new column in the summary data frame using Pandas to calculate count of purchases grouped by gender
gender_calc['Purchase Count'] = purchase_data.groupby('Gender')['Gender'].count()
# Create a new column in the summary data frame using Pandas to calculate the average purchase price by gender
gender_calc['Average Purchase Price'] = purchase_data.groupby('Gender')['Price'].mean()
# Create a new column in the summary data frame using Pandas to calculate the total purchase value by gender
gender_calc['Total Purchase Value'] = purchase_data.groupby('Gender')['Price'].sum()
# Create a new column in the summary data frame using Pandas to calculate the total number of unique SN's by gender
gender_calc['Unique SN Count'] = unique_sn.groupby('Gender')['SN'].count()
# Create a new column in the summary data frame calculating the avg total purchase by gender
gender_calc['Avg Total Purchase Per Person'] = gender_calc['Total Purchase Value']/gender_calc['Unique SN Count']
# Use map to format the currency columns
gender_calc['Average Purchase Price'] = gender_calc['Average Purchase Price'].map("${:.2f}".format)
gender_calc['Total Purchase Value'] = gender_calc['Total Purchase Value'].map("${:.2f}".format)
gender_calc['Avg Total Purchase Per Person'] = gender_calc['Avg Total Purchase Per Person'].map("${:.2f}".format)
# Remove the unique SN count column to clean up the summary data frame
gender_calc = gender_calc.drop(columns = ['Unique SN Count'])
# Sort the dataframe by the Gender column
gender_calc = gender_calc.sort_values(by=['Gender'], ascending = True)
# show the summary
gender_calc
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Create bins in which to place values based on Age column
bins = [0, 9, 14, 19, 24, 29, 34, 39, 100]
# Create labels for the bins
group_labels = ["< 10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
# Slice the Age data into the bins
age_group = pd.cut(unique_sn['Age'], bins, labels = group_labels)
# Add the age group data series to the DataFrame
unique_sn['Age Group'] = age_group
unique_sn.head()
```

```python
# Create a dataframe from the total counts of users by age group
age_sum = unique_sn['Age Group'].value_counts().rename_axis('Age Group').reset_index(name='Total Count')
# Add a column to the dataframe of the calculated percentage of users by age group
age_sum['Percentage of Players'] = age_sum['Total Count']/age_sum['Total Count'].sum()*100
# Reset the index to the Age Group column
age_sum = age_sum.set_index('Age Group')
# Convert the Percentage of Players column from string to numeric
age_sum['Percentage of Players'] = pd.to_numeric(age_sum['Percentage of Players'])
# Format the Percentage of Players column to show as a percent
age_sum['Percentage of Players'] = pd.Series(["{0:.2f}%".format(val) for val in age_sum['Percentage of Players']], index=age_sum.index)
# Sort the dataframe by Age Group
age_sum = age_sum.sort_values(by=['Age Group'], ascending = True)
# show the summary Data Frame
age_sum
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Isolate columns Item Name, Item ID and Price
items_df = purchase_data[[ 'Item Name','Item ID', 'Price']]
# Group rows by Item Name and Item ID
pop_items = items_df.groupby(['Item ID','Item Name']).count()
# Reset the index
pop_items = pop_items.reset_index()
# Calculate the number of purchases per item and add to a list to remove index
purchase_count= purchase_data.groupby(['Item ID','Item Name'])['Purchase ID'].count().tolist()
# Add the generated list to the summary DataFrame as a new column
pop_items["Purchase Count"] = purchase_count
# Calculate the average per item and add to a list to remove index
average_price = purchase_data.groupby(['Item ID','Item Name'])['Price'].mean().tolist()
# Add the generated list to the summary DataFrame as a new column
pop_items["Average Price"] = average_price
# Calculate the total purchase value per item and add to a list to remove index
purchase_value = purchase_data.groupby(['Item ID','Item Name'])['Price'].sum().tolist()
# Add the generated list to the summary DataFrame as a new column
pop_items['Total Purchase Value'] = purchase_value
# Remove original Price column to clean up summary DataFrame
pop_items.pop('Price')
# Sort the summary DataFrame by Purhcase count
pop_items = pop_items.sort_values(by=['Purchase Count'], ascending = False)
# Format the currency columns
pop_items['Average Price'] = pop_items['Average Price'].map("${:.2f}".format)
pop_items['Total Purchase Value'] = pop_items['Total Purchase Value'].map("${:.2f}".format)
# Set the index
pop_items = pop_items.set_index(['Item ID'], ['Item Name'])
pop_items.head()
```
